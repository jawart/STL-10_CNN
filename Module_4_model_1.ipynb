## 1. New model configuration

### 1.1 Existing model loading
- Available models:

```python
%%bash
ls ./Models
```

```python
model_0 = keras.models.load_model('./Models/model_0.h5') # basic model before training
model_0.name = 'model_0'
model_0.summary()
```

### 1.2 Extraction of convolutional layers from existing model 

```python
for i in range (0,10):
    print('Layer', i,' : ', model_0.layers[i].name)
```

```python
conv_base_0 = Model(inputs = model_0.input, outputs = model_0.get_layer(index = 9).output)
conv_base_0.name = 'conv_base_0'
conv_base_0.summary()
conv_base_0.save('./Models/conv_base_0.h5')
```

### 1.3 Extraction and update of the classifier

```python
classifier_0 = models.Sequential()
classifier_0.add(layers.Dense(512, activation='relu', input_dim=1*1*512))
classifier_0.add(layers.Dense(2, activation='sigmoid'))
classifier_0.name = 'classifier_0'
classifier_0.summary()
classifier_0.save('./Models/classifier_0.h5')
```

#### New 'Dropout' layer in the classifier:

```python
classifier_1 = models.Sequential()
classifier_1.add(layers.Dense(512, activation='relu', input_dim=1*1*512))
classifier_1.add(layers.Dropout(0.5))
classifier_1.add(layers.Dense(2, activation='sigmoid'))
classifier_1.name = 'classifier_1'
classifier_1.summary()
classifier_1.save('./Models/classifier_1.h5')
```

### 1.4 model_1 aggregation

```python
model_1 = models.Sequential()
model_1.add(conv_base_0)
model_1.add(layers.Flatten())
model_1.add(classifier_1)
model_1.name = 'model_1'
model_1.summary()
model_1.save('./Models/model_1.h5') # not trained
```
